```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
```

## Linear Regression model

```python
X = tf.constant([-10.0,-5.0, 0.0,5.0,10.0,15.0], dtype = np.float16)
Y = tf.constant([-6.0,-3.0,0,3,6,9], dtype = np.float16)

plt.scatter(X,Y)
plt.show()
```

```python
tf.random.set_seed(42)

model = tf.keras.Sequential([
    tf.keras.layers.Dense(1)

])

model.compile(loss =tf.keras.losses.MeanAbsoluteError,
              optimizer = tf.keras.optimizers.SGD(),
              metrics = ["mae"])

model.fit(tf.expand_dims(X, axis =-1),Y, epochs = 20)
```

```python
model.predict(tf.constant([20.0]))
```

## Bigger model

```python
model2 = tf.keras.Sequential([
    tf.keras.layers.Dense(100, activation ="relu"),
    tf.keras.layers.Dense(1)
])

model2.compile(loss =tf.keras.losses.MeanAbsoluteError,
              optimizer = tf.keras.optimizers.SGD(),
              metrics = ["mae"])

model2.fit(tf.expand_dims(X, axis =-1),Y, epochs = 10)
```

```python
model2.predict(tf.constant([20.0]))
```

## Bigger Dataset

```python
X2 = tf.range(-100,100,4)
Y2 = X2 + 10

plt.scatter(X2,Y2)
```

```python
len(X2)
```

```python
# creating train and test set
train_X, train_Y = X2[:40], Y2[:40]
test_X, test_Y = X2[40:], Y2[40:]
```

```python
#visualising data

plt.scatter(train_X,train_Y, label = "training data")
plt.scatter(test_X,test_Y, label = "testing data")
plt.legend()
plt.show()
```

```python
# create model 3

model3 = tf.keras.Sequential([
    tf.keras.layers.Dense(100, name = "Input_layer"),
    tf.keras.layers.Dense(1, name = "output_layer")
])

model3.compile(loss= tf.keras.losses.MeanAbsoluteError(),
              optimizer = tf.keras.optimizers.SGD(),
              metrics = ["mae"])

model3.fit(tf.expand_dims(train_X, axis = -1), train_Y, epochs = 20)
```

```python
model3.predict(test_X)
```

```python
model3.summary()
```

## Visualising NN

```python
from tensorflow.keras.utils import plot_model
plot_model(model3, show_shapes = True)
```

## Visualise predictions

```python
def plot_predictions(test_data = test_X,
                     test_labels = test_Y,
                     y_preds = None):
    """
    Plots training data, test data and compares predictions
    """


    plt.scatter(test_X, test_Y, label = "actual")
    plt.scatter(test_X, y_preds, label = "predicted")
    plt.legend()
    plt.show()
```

```python
y_preds = model3.predict(test_X)

plot_predictions(test_X, test_Y, y_preds)
```

## Ficing model 3

```python
model3_MK2 = tf.keras.Sequential([
    tf.keras.layers.Dense(100, name = "Input_layer"),
    tf.keras.layers.Dense(10, name = "Input_layer_2"),
    tf.keras.layers.Dense(1, name = "output_layer")
])

model3_MK2.compile(loss= tf.keras.losses.MeanAbsoluteError(),
              optimizer = tf.keras.optimizers.SGD(learning_rate = 0.003),
              metrics = ["mae"])

model3_MK2.fit(tf.expand_dims(train_X, axis = -1), train_Y, epochs = 25)
```

```python
y_preds = model3_MK2.predict(test_X)

plot_predictions(test_X, test_Y, y_preds)
```

## Running experiments

```python
tf.random.set_seed(42)
```

```python
import pandas as pd
def experiment(epoch,
               metrics,
               layer1_neurons,
               layer2_neurons,
               loss,
               ):

  results = {"model name":[],
             "model MAE" :[],
             "model MSE" :[]}

  for x in range(len(layer1_neurons)):
    for y in range(len(layer2_neurons)):
      for z in range(len(epoch)):
        model = tf.keras.Sequential([
          tf.keras.layers.Dense(layer1_neurons[x]),
          tf.keras.layers.Dense(layer2_neurons[y]),
          tf.keras.layers.Dense(1)
        ])



        name = str(layer1_neurons[x]) + "_" + str(layer2_neurons[y])+"_"+str(epoch[z])+"_"+str(loss[x])

        model.compile(loss = loss[x],
                      optimizer = tf.keras.optimizers.SGD(learning_rate = 0.003),
                      metrics = metrics)

        model.fit(tf.expand_dims(train_X, axis = -1), train_Y, epochs = epoch[z], verbose = 1)

        y_preds = model.predict(test_X)

        results["model name"].append(name)
        results["model MAE"].append(mae(test_Y, y_preds).numpy())
        results["model MSE"].append(mse(test_Y, y_preds).numpy())

        plot_predictions(test_X, test_Y, y_preds)
        tf.keras.backend.clear_session()



  return pd.DataFrame(results)


```

```python
results = experiment(
    [10,25,50],
    ["mae"],
    [200,100,50],
    [20,15,10],
    ["mae", "mae", "mae"]
)
```

## Saving model

```python
best_model = tf.keras.Sequential([
    tf.keras.layers.Dense(50, name = "Input_layer"),
    tf.keras.layers.Dense(10, name = "Input_layer_2"),
    tf.keras.layers.Dense(1, name = "output_layer")
])

best_model.compile(loss= tf.keras.losses.MeanAbsoluteError(),
              optimizer = tf.keras.optimizers.SGD(learning_rate = 0.003),
              metrics = ["mae"])

best_model.fit(tf.expand_dims(train_X, axis = -1), train_Y, epochs = 50)
```

```python
Y_pred = best_model.predict(test_X)
#Y_pred, test_Y
mae(Y_pred, test_Y).numpy()
```
