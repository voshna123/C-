```python
import tensorflow as tf
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import sklearn
```

```python
df1 = tf.range(-100,100,4)
y = df1+10

df1,y
```

```python
train_X , test_X , train_y , test_y = sklearn.model_selection.train_test_split(pd.DataFrame(df1),pd.DataFrame(y),test_size=0.2)
```

```python
model1= tf.keras.Sequential([
    tf.keras.layers.Dense(100,input_shape=[1]),
    tf.keras.layers.Dense(50),
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])

model1.compile(
    loss = tf.keras.losses.mae,
    optimizer = tf.keras.optimizers.Adam(),
    metrics = ["mae"]
)

model1.fit(
    train_X,train_y,
    epochs=300,
)
```

```python
model1.evaluate(test_X,test_y)
```

## Boston Housing dataset

### Loading data and creating train and test datasets

```python
train,test =tf.keras.datasets.boston_housing.load_data(
    path='boston_housing.npz', test_split=0.2, seed=113
)

train_X , test_X , train_y , test_y = pd.DataFrame(train[0]) , pd.DataFrame(test[0]) , pd.DataFrame(train[1]) , pd.DataFrame(test[1])
```

```python
train_X.shape
```

```python
model2= tf.keras.Sequential([
    tf.keras.layers.Dense(100),
    tf.keras.layers.Dense(50),
    tf.keras.layers.Dense(10),
    tf.keras.layers.Dense(1)
])

model2.compile(
    loss = tf.keras.losses.mae,
    optimizer = tf.keras.optimizers.Adam(),
    metrics = ["mae"]
)

hist = model2.fit(
    train_X,train_y,
    epochs=300,
)
```

```python
model2.evaluate(test_X,test_y)
```

```python
pd.DataFrame(hist.history).plot()
plt.xlabel("Epochs")
plt.ylabel("Loss")
plt.title("Loss Curve")
```
